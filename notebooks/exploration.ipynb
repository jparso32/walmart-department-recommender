```python
features['MarkDown1'] = features['MarkDown1'].fillna(0)
features['MarkDown2'] = features['MarkDown2'].fillna(0)
features['MarkDown3'] = features['MarkDown3'].fillna(0)
features['MarkDown4'] = features['MarkDown4'].fillna(0)
features['MarkDown5'] = features['MarkDown5'].fillna(0)

features.isnull().sum()
# CPI and Uneployment had missing values, so we filled them with the median for both. CPI and Unemployment are continuous variables, and making them 0 would not make sense economically. 

features['CPI'] = features['CPI'].fillna(features['CPI'].median())
features['Unemployment'] = features['Unemployment'].fillna(features['Unemployment'].median())
```

```python
#We need to make the data column into a datetime object so we can extract the month and year from it.
train['Date'] = pd.to_datetime(train['Date'])
features['Date'] = pd.to_datetime(features['Date'])
```

```python
#We are now merging the 2 dataframes together

features_clean = features.drop(columns=['IsHoliday'])
data = train.merge(features_clean, on=['Store', 'Date'], how='left')
data = data.merge(stores, on='Store', how='left')

data.head()
```

```python
# We want to extract the week, month, and year to see if there are any trends, the model can't understand a date so we have to change the date into something it can understand.
# so we make 3 new columns to signify the differences. isocalendar helps us find the week of the year using a calendar. 

data['Week'] = data['Date'].dt.isocalendar().week.astype('int')
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

data.head()
```

```python
# Adjusting the type of store column so its easier to work with. map goes through every row.
# Type A = The Largest stores, highest sales volume
# Type B = Medium sized stores, moderate sales volume
# Type C = Smallest stores, lowest sales volume

data['Type'] = data['Type'].map({'A': 0, 'B': 1, 'C': 2})
```

```python
#Checked the columns because some data handling didn't work, checked and saw there were 2 IsHoliday's. I am changing the feature to only one IsHoliday column.

#data.columns.tolist()

#data = data.drop(columns=['IsHoliday_y'])
#data = data.rename(columns={'IsHoliday_x': 'IsHoliday'})
#data["IsHoliday"] = data['IsHoliday'].astype(int)
```

```python
# Handling IsHoliday column, we will make it 1 for true and 0 for false, this way the model can understand it better.
data['IsHoliday'] = data['IsHoliday'].map({True: 1, False: 0})
data.head()
```
